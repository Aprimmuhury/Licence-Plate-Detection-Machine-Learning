```python
# Step 1: Required Libraries Install
!pip install matplotlib pillow

# Step 2: Libraries Import
import os
from google.colab import files
from PIL import Image
import matplotlib.pyplot as plt

# Step 3: Upload Images
print("Please select your 10 JPG images...")
uploaded = files.upload()

# Step 4: Create a Folder for Images
os.makedirs("test_images", exist_ok=True)

# Save uploaded images in folder
for filename, filedata in uploaded.items():
    with open(os.path.join("test_images", filename), "wb") as f:
        f.write(filedata)

print(f"Uploaded {len(uploaded)} images successfully!")

# Step 5: Display All Images
image_files = sorted(os.listdir("test_images"))

plt.figure(figsize=(15, 8))
for i, img_name in enumerate(image_files):
    img_path = os.path.join("test_images", img_name)
    img = Image.open(img_path)
    plt.subplot(2, 5, i + 1)
    plt.imshow(img)
    plt.title(img_name)
    plt.axis("off")
plt.tight_layout()
plt.show()
```
